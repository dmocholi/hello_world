```python
import csv 

### READ THE .csv FILE 
f = open('PitchingPost.csv')
data = csv.reader(f)

# Extract the years and the freq. of this years
year = {} # Initilize dict
for line in data:
    if line[1] in year.keys():
        year[line[1]] +=  1 # Each time we repeat a year add 1 to the fre. counter
        
    else:
        year[line[1]] = 1 # Initilize the data for this key year

### WRITE THE NEW .csv FILE

f_ex = open('Exit.csv', 'w')
exit = csv.writer(f_ex)

for i in year.keys():
    string = [str(i), str(year[i])]
    exit.writerow(string)

f.close()
f_ex.close()
    


```

Crear una serie denominada "Sueldos_hombres" que contenga los datos [2500,1800,1900,2000,2100] y cuyos ndices sean ["Euskadi", "Murcia", "Madrid", "Barcelona", "Zaragoza"], y otro denominada "Sueldos_mujeres" que contenga los datos [2300,1600,1980,1900,2150] y cuyos ndices sean ["Euskadi", "Murcia", "Madrid", "Barcelona","Crdoba"]
Usando las series anteriores, obtn otra serie que contenga la diferencia de sueldos entre mujeres y hombres.
Usando la serie anterior que contiene las diferencias de sueldos, obtn las ciudades donde la diferencia de sueldos es mayor de 100 .


```python
import pandas as pd

Sueldos_hombres = pd.Series([2500,1800,1900,2000,2100], index = ["Euskadi", "Murcia", "Madrid", "Barcelona", "Zaragoza"])
Sueldos_mujeres = pd.Series([2300,1600,1980,1900,2150], index = ["Euskadi", "Murcia", "Madrid", "Barcelona", "Crdoba"])

Sueldos_dif = Sueldos_hombres - Sueldos_mujeres


mask = abs(Sueldos_dif) > 100
mask
```

```python
Sueldos_dif.values
```

```python
ciudades = Sueldos_dif[mask]
print(ciudades)
```
