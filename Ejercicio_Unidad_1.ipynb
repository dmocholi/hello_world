```python
def match_value(words_dict,text):
    
    ### Filtra las palabras del texto (filter_words)
    text_words = text.split() # Separa el texto en palabras
    characters = ['.', ',', '!', '?', '¿']
    text_words = filter_words(text_words,characters) # OUTPUT: Lista de palabras del tweet filtradas
    
    ### Calcula el valor del sentimiento del tweet
    value = 0 # Inicializa el valor de sentimiento de cada tweet
    match_words = [] # Inicializa una lista para guardar las palabras que llevan asociado un valor
    for item in words_dict.keys(): # Recorre todos los items en el archivo para ver si se encuentran en el texto del tweet
        # Separa cada item en palabras
        item_word = item.split()
        
        # Si el item esta compuesto de mas de una palabra entra aqui
        if len(item_word) > 1:
            if item_word[0] in text_words: # Entra si la primera palabra del item esta en la lista de palabras del tweet 
                match = True # Asigna una etiqueta
                pos = text_words.index(item_word[0]) # Calcula la posicion donde se encuentra en la lista de palabras del tweet
                
                # Loop para ver si las sig. palabras del item estan en las pos. contiguas de la lista de palabras del tweet 
                for loc in range(len(item_word))[1:]: 
                    if item_word[loc] == text_words[pos + 1]:
                        pos = text_words.index(item_word[loc]) # Actualiza la posicion
                    else:
                        match = False # Cambia la etiqueta
                        
                if match == True: # Si la etiqueta es igual a True hay conincidencia
                    value += words_dict[item] # Se suma el valor del item a la variable que guarda el valor del tweet
                    match_words.append(item) # Se guarda la plabra coincidente
        
        # Si el item esta compuesto solo por una palabra entra aqui
        elif len(item_word) == 1:
            if item in text_words: # Si el item se encuentra en la lista de palabras del tweet
                value += words_dict[item] # Se suma el valor del item a la variable que guarda el valor del tweet
                match_words.append(item) # Se guarda la plabra coincidente
    
    return value, match_words # Devuelve las variables
    

def match_value_re(words_dict,text):
    
    ### Calcula el valor del sentimiento del tweet
    value = 0 # Inicializa el valor de cada tweet
    match_words = [] # Inicializa una lista para guardar las palabras que llevan asociado un valor
    for item in words_dict.keys(): # Recorre todos los items en el archivo para ver si se encuentran en el texto del tweet
        if re.search(r'\b' + item + r'\b', text): # Con regular expressions comprueba si item se encuentra en texto del tweet
            value += words_dict[item] # Se suma el valor del item a la variable que guarda el valor del tweet
            match_words.append(item) # Se guarda la plabra coincidente
        else:
            pass
        
    return value, match_words # Devuelve las variables


def tweet_feeling(tweets_file, feelings_file):
    
    ### Abrir los archivos de texto
    f1 = open(tweets_file)
    f2 = open(feelings_file)
    
    ### Crear un dict para almacenar cada item (palabra o conjunto palabras) junto su valor de sentimiento
    feelings = {} # Inicializa el dict
    for line in f2:
        item_feeling,value_feeling = line.split('\t')
        feelings[item_feeling] = int(value_feeling)
    
    ### Iterar dentro de cada tweet almacenado en tweets_file
    data_tweets = {} # Inicializa el dict donde se almacenara todas la info referente al tweet procesado
    cont = 1 # Contador de tweets
    for line in f1:
        tweet = json.loads(line) # Carga cada tweet (Se encuentran en formato json)
        if 'text' in tweet.keys(): # Solo procesa el tweet si contiene texto
            data_tweets['Tweet ' + str(cont)] = {} # Inicializo un dict para cada tweet procesado para guardar la info. 
            text_tweet = tweet['text'].lower() # Convertir el texto a minusculas

            ### Aplicamos la funcion de comparación
            tweet_value, match_words = match_value(feelings, text_tweet)
            #tweet_value, match_words = match_value_re(feelings, text_tweet)

            # Guarda la info. obtenida en el dict del tweet
            data_tweets['Tweet ' + str(cont)]['text'] = tweet['text'] 
            data_tweets['Tweet ' + str(cont)]['valor'] = tweet_value
            data_tweets['Tweet ' + str(cont)]['palabras'] = match_words

            # Imprimimos la info. por pantalla
            #print("EL SIGUIENTE TWEET: " + "\"" + tweet['text'] + "\" " + "TIENE UN SENTIMIENTO ASOCIADO DE: " + str(tweet_value))

        else:
            pass
        cont += 1 # Suma una iteracion al contador
    
    return data_tweets # Devuelve el dict con la info. de cada tweet


def filter_words(words, characters):
    
    ### Filtra una lista de palabras
    filttered_words = [] # Inicializa la lista donde almacenar las palabras filtradas
    for i in words:
        temp_word = i # Asigna un nombre temporal
        
        # Primer filtro: quitar caracteres
        for c in characters:
            if i.endswith(c) or i.startswith(c):
                temp_word = i.replace(c,'')
            
        # Segundo filtro: chequear que es contenido alfabetico
        if temp_word.isalpha():
            filttered_words.append(temp_word)
        else:
            pass
        
    return filttered_words # Devuelve la lista con las palabras filtradas
```
